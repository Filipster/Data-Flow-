```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```
import folium
```

```
data = pd.read_csv('convertcsv (1).csv')
```

```
data.sample(5)
```

```
data['colAcute'] = data.apply(lambda x: colPick(data), axis=1)
```

```
def colPick():
  for i in data['Acutedisease_or_seroprevalence']:
    if i == 'Acute':
      return ('Red')
    else:
      return ('Blue')
```

```
print(max(data.Latitude))
print(max(data.Longitude))
print(min(data.Latitude))
print(min(data.Longitude))
```

```
mp = folium.Map([17.9, 14.4], zoom_start=4.5) #range from open street Map
#mp
```

```
for index, row in data.iterrows():
  folium.CircleMarker([row['Latitude'], row['Longitude']],
                        radius=4,
                        fill = True,
                        opacity = row['Number_acutecases'],
                        fill_color = row['colAcute'],
                        stroke=True,
                        popup=row['Town'],
                        color= row['colAcute'],
                        ).add_to(mp)

mp
```

```
from folium.plugins import HeatMap
data['count'] = 1
HeatMap(data=data[['Latitude','Longitude','count']].groupby(['Latitude', 'Longitude']).sum().reset_index().values.tolist(), radius=8, max_zoom=13).add_to(mp)

mp
```
